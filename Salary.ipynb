```python
import pandas as pd
df=pd.read_csv("salaries.csv")
```

```python
df
```

```python
print(df['company'].value_counts())
print(df['job'].value_counts())
print(df['degree'].value_counts())
```

```python
df['company']=df['company'].replace({'google':0,'abc pharma':1,'facebook':2})
df['job']=df['job'].replace({'business manager':0,'sales executive':1,'computer programmer':2})
df['degree']=df['degree'].replace({'bachelors':0,'masters':1})
```

```python
df
```

```python
print(df.corr().to_string())
```

```python
x=df.drop(columns=['salary_more_then_100k','job'])
y=df['salary_more_then_100k']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

### LOGISTIC RGRESSION

```python
from sklearn.linear_model import LogisticRegression
modellog=LogisticRegression()
```

```python
modellog.fit(x,y)
```

```python
modellog.score(x,y)
```

```python
modellog.fit(x_train,y_train)
```

```python
modellog.score(x_train,y_train)
```

### DECISION TREE


```python
from sklearn.tree import DecisionTreeClassifier
modeldec=DecisionTreeClassifier()
modeldec.fit(x,y)
```

```python
modeldec.score(x,y)
```

```python
modeldec.fit(x_train,y_train)
```

```python
modeldec.score(x_test,y_test)
```
